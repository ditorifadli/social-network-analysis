```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/timnas_indonesia_tweets (1).csv')

# Create nodes dataframe
nodes = pd.DataFrame({
    'Id': df['user_id_str'],
    'Label': df['username'],
    'Location': df['location']
}).drop_duplicates()

# Create edges dataframe
edges = df[df['in_reply_to_screen_name'].notnull()]
edges = pd.DataFrame({
    'Source': df['user_id_str'],  # User who tweeted
    'Target': df['in_reply_to_screen_name'],  # User who is replied to
    'Type': 'Directed',
    'Weight': 1  # Set to 1 for each interaction
})

# Save nodes and edges to CSV for Gephi
nodes.to_csv('gephi_nodes.csv', index=False)
edges.to_csv('gephi_edges.csv', index=False)
```
